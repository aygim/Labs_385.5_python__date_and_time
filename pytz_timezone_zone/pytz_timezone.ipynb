# Time Zone Conversion w/ Pytz


1.install pytx: pip install pytz

```python
import pytz
from datetime import datetime
```

```python
# create a function with pytz to convert timezones
#Accepts 3 parameters:
# dt - the date/time
# from tz - for current tz
# to_tz - timezone to convert time too
def convert_time_xone(dt,from_tz, to_tz):

    #create timezone objects
    # taking tz string and creating special tz oblkect with pytz
    from_tz = pytz.timezone(from_tz)
    to_tz = pytz.timezone(to_tz)

    #Convert/add the tz to the datetime object if necessary
    #setting the tz attribute
    dt = dt.replace(tzinfo=from_tz)

# converting dateime to new timezone
    dt = dt.astimezone(to_tz)
    
    return dt
```

```python
#Using created function
today_datetime = datetime.now()

from_timezone = 'US/Eastern'
to_timezone ='Australia/Melbourne'

converted_time = convert_time_xone(today_datetime, from_timezone, to_timezone)

print(converted_time)
```
