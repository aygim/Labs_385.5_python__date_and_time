# Date Time Overview


```python
#importing datetime module
import datetime
from datetime import date, datetime, time
```

```python
# Creating a time object
todaysDate = date(2026, 4, 29)
print(todaysDate)
```

```python
lastweek = date(2026, 4, 22)
print(lastweek)
```

```python
print(date.today())
```

```python
today = date.today()
print(today)
```

```python
#You can subtract dates using datetime.
print(today - lastweek)
```

```python
# Datetime gives both date and time
right_now = datetime.now()
print(right_now)
```

```python
# With dates build with datetime module you get direct access
print(right_now.year)
print(right_now.month)
print(right_now.day)
print(right_now.minute)
print(right_now.second)
print(right_now.microsecond)
```

```python
time_right_now = time(21, 0, 0)
print(time_right_now)
print(time_right_now.hour)
```

```python
# strftime() - string format time - reformats time as a string
print(date.today().strftime('%A %B %d %Y '))
```
